# Data Visualization of Manufacturing Defects

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Explore Data

```python
df = pd.read_csv('../test_data/manufacturing_defects.csv')
print('shape:', df.shape, end='\n\n')
print(df.dtypes, end='\n\n')
df.head()
```

### Check Value Counts 
Data is very imbalanced

```python
# 0 for low defect; 1 for high defect
df['DefectStatus'].value_counts()
```

```python
df['Bin'] = (df['ProductionVolume'] // 100) * 100
data = df.groupby(['Bin','DefectStatus']).size().unstack(fill_value=0)

data.plot(kind='bar', stacked=False, figsize=(10,6), color=['skyblue','orange'])
plt.title('High Defect Count by Production Volume')
plt.xlabel('Production Volume By 100s')
plt.ylabel('Number of High Defects')
plt.legend(title='Defect Status', labels=['Low', 'High'])
plt.xticks(rotation=0)
plt.show()
```

```python
df = df.drop('Bin', axis=1)
```

### Determining Correlation 

```python
plt.figure(figsize=(16,12))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
most_corr = df.corr()['DefectStatus'].abs().sort_values(ascending=False)
most_corr
```

```python
most_corr = most_corr[most_corr >= 0.05].index.tolist()
sns.pairplot(df[most_corr], hue='DefectStatus')
```
